## Importance of data being normal and stationary

T-tests require that our data is normality distributed. 

The tests that we use to validate our models require that the data be stationary over time. 

Data being stationary over time means that the mean, variance and covariance of the data is the time over time. 

Variance being constant over time is called homoscedasticity

Variance changing over time is called hetroscedasticity. 

We use the Breusch-Pagan test to see if our data is heteroscedastic.

## Autoregessive (AR) Model

Assume that the previous value gives us some insight in to the future value. 

$ y_t = \alpha + B_1 y_{t-1} + B_2 y_{t-2}  + ... \epsilon_t $

AR models are defined by the lag. 

A model that uses only the previous one day's value is an AR(1) model. 2 days AR(2). We use AR(p) where p represents the lag. 

If some coefficients are greater than 0 then they explain they explain some of the future predict. 

We can also use regression to see the $R^2$ value to see how well the independent variables explain the dependent  variables.
