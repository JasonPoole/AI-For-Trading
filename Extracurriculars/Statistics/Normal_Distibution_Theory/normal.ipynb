```python
import numpy as np
import matplotlib.pyplot as plt
```

Simple 20 flips of a fair coin 1 million times plotted

```python
flips, prob, trails = 20, 0.5, int(1e6)
ten_flips = np.random.binomial(flips, prob, trails)
plt.hist(ten_flips, alpha=0.4, histtype='bar', ec='black', bins=20, color='purple')
plt.title(f'number of heads after {flips} flips')
plt.ylabel('frequency')
plt.xlabel('number of heads')
plt.show()
```
