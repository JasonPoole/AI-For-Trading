```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams['figure.figsize'] = (16,8)
```

```python
BTC_price_data = yf.download('BTC-USD','2017-01-01','2021-01-01')
google_price_data = yf.download('GOOGL','2017-01-01','2021-01-01') 
```

```python
BTC_price_data['Adj Close'].plot(color='orange')
google_price_data['Adj Close'].plot(color='blue')
plt.title('BTC v Google')
plt.ylabel('Price')
plt.grid()
plt.yscale('log')
plt.legend(['BTC', 'Google']);
plt.show()
```
